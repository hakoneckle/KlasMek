```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
%matplotlib inline
```

```python
# Constants
m1 = 2000                        # counterweight
m2 = 15                          # Projectile
l1 = 1.2
l2 = 5.7
l3 = 3.2
l4 = 5
g = 9.81
theta0 = 0.7*np.pi               # initial angle theta
theta0_der = 0                   # initial change in angle theta
psi0 = theta0 - 0.5*np.pi        # initial angle psi
psi0_der = 0                     # initial change in angle psi
```

### Task 1:



First, I will give the expressions for the velocities of countermass ($\vec{v_1}$) and the projectile ($\vec{v_{2b}}$): 

$$\vec{v_1} = \dot{\vec{r_1}} = 
\frac{d}{dt} \begin{pmatrix} 
l_1 sin(\theta) \\ 
-l_1 cos(\theta) + l_3 
\end{pmatrix} = 
\begin{pmatrix} 
l_1 \dot{\theta} cos(\theta) \\ 
l_1 \dot{\theta} sin(\theta) 
\end{pmatrix}$$

$$\vec{v_{2b}} = \dot{\vec{r_{2b}}} = 
\frac{d}{dt} \begin{pmatrix} 
-l_2 sin(\theta) -l_4 sin(\psi-\theta) \\ 
l_2 cos(\theta) + l_3 - l_4 cos(\psi - \theta) 
\end{pmatrix} = 
\begin{pmatrix} 
-l_2 \dot{\theta} cos(\theta) -l_4 (\dot{\psi} - \dot{\theta}) cos(\psi-\theta) \\ 
-l_2 \dot{\theta} sin(\theta) + l_4 (\dot{\psi} - \dot{\theta}) sin(\psi - \theta)
\end{pmatrix}$$

The Lagrangian for the system is:

$$L(\theta,\dot{\theta},\psi,\dot{\psi}) = \sum_{i}T_i - V_i = \frac{1}{2}m_1\vec{v_1}^2 - m_1gr_{1y} + \frac{1}{2}m_2\vec{v_{2b}}^2 - m_2gr_{2y}$$

where $\vec{v_1}$ and $\vec{v_{2b}}$ is as above, and $r_{1y}$ and $r_{2y}$ are the $y$-components of the position vector. This gives us

$$
L(\theta,\dot{\theta},\psi,\dot{\psi}) = 
\frac{1}{2}m_1(l_1^2\dot{\theta}^2) + 
\frac{1}{2}m_2(-l_2 \dot{\theta} cos(\theta) -l_4 (\dot{\psi} - \dot{\theta}) cos(\psi-\theta))^2 + 
\frac{1}{2}m_2(-l_2 \dot{\theta} sin(\theta) + l_4 (\dot{\psi} - \dot{\theta}) sin(\psi - \theta))^2 
-m_1g(-l_1 cos(\theta) + l_3)
-m_2g(l_2 cos(\theta) + l_3 - l_4 cos(\psi - \theta)) \\ 
=
\frac{1}{2}m_1l_1^2\dot{\theta}^2 +
\frac{1}{2}m_2(l_2^2 \dot{\theta}^2 cos^2(\theta) + l_4^2 (\dot{\psi}^2 + \dot{\theta}^2 - 2\dot{\psi}\dot{\theta}) cos^2(\psi-\theta) + 2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})cos(\theta)cos(\psi-\theta)) + 
\frac{1}{2}m_2(l_2^2 \dot{\theta}^2 sin^2(\theta) + l_4^2 (\dot{\psi}^2 + \dot{\theta}^2 - 2\dot{\psi}\dot{\theta}) sin^2(\psi-\theta) - 2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})sin(\theta)sin(\psi-\theta)) +
-m_1g(-l_1 cos(\theta) + l_3)
-m_2g(l_2 cos(\theta) + l_3 - l_4 cos(\psi - \theta)) \\ 
= 
\frac{1}{2}(m_1l_1^2 + m_2l_2^2 cos^2(\theta) + m_2l_2^2 sin^2(\theta))\dot{\theta}^2 +
\frac{1}{2}m_2l_4^2 (\dot{\psi}^2 + \dot{\theta}^2 - 2\dot{\psi}\dot{\theta}) cos^2(\psi-\theta) +
m_2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})cos(\theta)cos(\psi-\theta)) +
\frac{1}{2}m_2l_4^2 (\dot{\psi}^2 + \dot{\theta}^2 - 2\dot{\psi}\dot{\theta}) sin^2(\psi-\theta) -
m_2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})sin(\theta)sin(\psi-\theta)) -
m_1g(-l_1 cos(\theta) + l_3) -
m_2g(l_2 cos(\theta) + l_3 - l_4 cos(\psi - \theta)) \\ 
=
\frac{1}{2}(m_1l_1^2 + m_2l_2^2)\dot{\theta}^2 + 
\frac{1}{2}m_2l_4^2(\dot{\psi} - \dot{\theta})^2 +
m_2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})cos(\psi) +
g(m_1l_1 - m_2l_2)cos(\theta) -
gl_3(m_1+m_2) +
gm_2l_4cos(\psi-\theta)
$$

Now, we can use the Euler-Lagrange equations on the Lagrangian from above:

$$
\frac{\partial L}{\partial \theta} - \frac{\partial}{\partial t} \frac{\partial L}{\partial \dot{\theta}} = 0 \\ 
[-g(m_1l_1-m_2l_2)sin(\theta)+m_2gl_4sin(\psi-\theta)] - \frac{\partial}{\partial t} [(m_1l_1^2+m_2l_2^2)\dot{\theta}-m_2l_4^2(\dot{\psi}-\dot{\theta})+m_2l_2l_4cos(\psi)(\dot{\psi}-2\dot{\theta})] = 0 \\ 
\ddot{\theta}\cdot[-m_1l_1^2-m_2l_2^2-m_2l_4^2+2m_2l_2l_4cos(\psi)] + \ddot{\psi}\cdot[m_2l_4^2-m_2l_2l_4cos(\psi)] = g(m_1l_1-m_2l_2)sin(\theta)-gm_2l_4sin(\psi-\theta)-m_2l_2l_4(\dot{\psi}-2\dot{\theta})\dot{\psi}sin(\psi)
$$

and

$$
\frac{\partial L}{\partial \psi} - \frac{\partial}{\partial t} \frac{\partial L}{\partial \dot{\psi}} = 0 \\
[-m_2l_2l_4\dot{\theta}(\dot{\psi}-\dot{\theta})sin(\psi)-m_2gl_4sin(\psi-\theta)] - \frac{\partial}{\partial t} [m_2l_4^2(\dot{\psi}-\dot{\theta})+m_2l_2l_4\dot{\theta}cos(\psi)] = 0 \\ 
\ddot{\theta}\cdot[m_2l_4^2-m_2l_2l_4cos(\psi)] + \ddot{\psi}\cdot[-m_2l_4^2] = m_2l_4[gsin(\psi-\theta)-l_2\dot{\theta}^2sin(\psi)]
$$

From these we get equation (6) from the assignment text, with the correct values for the matrix elements.

### Task 2:

Here, we are going to solve the equations of motion of the system using the RK4-algorithm.

```python
def F(y):
    
    a11 = -(l1**2)*m1 - (l2**2)*m2 + 2*l2*l4*m2*np.cos(y[2]) - (l4**2)*m2
    a12 = -l4*m2*(l2*np.cos(y[2]) - l4)
    a22 = -(l4**2)*m2
    A = np.array([[a11 , a12] , [a12 , a22]])
    
    b1 = g*m1*l1*np.sin(y[0]) - g*m2*(l2*np.sin(y[0]) + l4*np.sin(y[2] - y[0])) - l2*l4*m2*(y[3]-2*y[1])*np.sin(y[2])*y[3]
    b2 = l4*m2*(g*np.sin(y[2]-y[0]) - l2*np.sin(y[2])*(y[1]**2))
    b = np.array([b1,b2])
    
    der = np.zeros(4)
    
    C = np.matmul(np.linalg.inv(A),b)     # A^(-1)*b
    
    der[0] = y[1]
    der[1] = C[0]
    der[2] = y[3]
    der[3] = C[1]
    
    return der

def init():              # create array with initial values
    y = np.zeros(4)
    y[0] = theta0
    y[1] = theta0_der
    y[2] = psi0
    y[3] = psi0_der
    return y
```

```python
def RK4(dt,T,F,ini):                         # RK4-algorithm
    time = np.arange(0,T,dt)
    vec = np.zeros([len(time),4])
    vec[0] = ini                             # Create an initital vector with the initial values
    for i in range(1,len(time)):
        K1 = F(vec[i-1])
        K2 = F(vec[i-1] + dt/2)
        K3 = F(vec[i-1] + dt/2)
        K4 = F(vec[i-1] + dt)
        
        vec[i] = vec[i-1] + dt/6 * (K1 + 2*K2 + 2*K3 + K4) # RK4
    
    return np.array(vec),time
```

```python
arr,t = RK4(0.001,5,F,init())

plt.figure(dpi=120)
plt.plot(t,arr[:,0],label=r'$\theta$')
plt.plot(t,arr[:,1],label=r'$\dot{\theta}$')
plt.plot(t,arr[:,2],label=r'$\psi$')
plt.plot(t,arr[:,3],label=r'$\dot{\psi}$')
plt.xlabel('Time [s]')
plt.legend()
plt.show()
```

In the plot above we can see the development in angles and their derivatives over a time period of 5 seconds. We can see that they are periodic. If we run the system for longer time it will become unstable.

```python
def traj(arr): # gives the x- and y-values for the trajectory
    return [-l2*np.sin(arr[:,0]) - l4*np.sin(arr[:,2] - arr[:,0]) , l2*np.cos(arr[:,0]) + l3 - l4*np.cos(arr[:,2] - arr[:,0])]

dev = traj(arr)
plt.figure(dpi=120)
plt.plot(dev[0],dev[1])
plt.show()
```

When we don't release the projectile, this is the motion it will follow. It's a shape that makes sense, as it will be attached to the arms that rotate around a point.

### Task 3:

Now, we are going to release the projectile at an angle $\theta_r$. We assume no air resistance.

```python
def F_noair(y): # calculate derivative, y = [x,v_x,y,v_y]
    der = np.zeros(4) # der = [x',v_x',y',v_y']
    der[0] = y[1]
    der[1] = 0
    der[2] = y[3]
    der[3] = -g
    return der

def find_close(theta_r,arr,lim): # Find the closest point to the release angle for releasing the projectile
    d = 10
    x = -10
    for i in range(len(arr[:,0])):
        if abs(arr[i,0] - theta_r) < lim:
            d = arr[i,0]-theta_r
            x = i
            return d,x,arr[x,0]  # d = distance from theta_r , x = point in array , arr[x,0] = angle

def init_release(vec):  # initial values when projectile is released
    arr = np.zeros(4)   # [x,vx,y,vy]
    arr[0] = -l2*np.sin(vec[0]) -l4*np.sin(vec[2]-vec[0])
    arr[1] = -l2*vec[1]*np.cos(vec[0]) -l4*(vec[3]-vec[1])*np.cos(vec[2]-vec[0])
    arr[2] = l2*np.cos(vec[0]) +l3 -l4*np.cos(vec[2]-vec[0])
    arr[3] = -l2*vec[1]*np.sin(vec[0]) +l4*(vec[3]-vec[1])*np.sin(vec[2]-vec[0])
    return arr

def trajectory(theta_r,lim,arr=arr):              # calculate trajectory
    diff,when,angle = find_close(theta_r,arr,lim)
    start = init_release(arr[when])
    sling = traj(arr[:when,:])
    coords,t = RK4(0.001,20,F_noair,start)
    dist = 100
    T0 = 0
    for i in range(len(coords[:,2])):
        if abs(coords[i,2]) < dist:
            dist = abs(coords[i,2])
            T0 = i
    return sling,coords,start,T0
   
sling,coords,start,T0 = trajectory(0.1*np.pi,0.001) 
plt.figure(dpi=120)
plt.title(r'Trajectory for $\theta_r=0.1\pi$')
plt.plot(sling[0],sling[1])
plt.plot(coords[:T0,0],coords[:T0,2])
plt.plot(start[0],start[2],'o')
plt.xlabel('Distance (x)')
plt.ylabel('Height (y)')
plt.grid()
plt.show()
```

The plot above shows the trajectory of the projectile when released on $\theta_r=0.1\pi$. The blue part of the plot is when it's still attached to the trebuchet, the green point is the release point, while the orange part is the trajectory after the release.

```python
angs = [0,0.05,0.1,0.15,0.2]
lims = [0.001,0.001,0.001,0.01,0.001]
for i in range(len(angs)):
    sling,coords,start,end = trajectory(angs[i]*np.pi,lims[i])
    plt.figure(dpi=120)
    plt.title(f'Release angle: {angs[i]}$\pi$')
    plt.plot(sling[0],sling[1])
    plt.plot(coords[:end,0],coords[:end,2])
    plt.plot(start[0],start[2],'o')
    plt.grid()
    plt.show()
```

We now plotted the trajectories for different angles in the interval ($0^\circ$,$40^\circ$). As we can see, the distances fired differs greatly. For $\theta_r=0.2\pi$ the trebuchet fires backwards. To see where the trebuchet changes from firing forwards to backwards, we plot the distance as a function of the release angle:

```python
distances = []
angles = np.linspace(0,0.4*np.pi,40)
turn = 0
for i in range(len(angles)):
    s,dis,star,en = trajectory(angles[i],0.01)
    distances.append(dis[en,0])
    if i>0:
        if distances[i]*distances[i-1] < 0:
            turn = i
print(f'The trebuchet fires in forward direction for release angles under {(angles[turn]+angles[turn-1])/(2*np.pi):.3f}\u03C0')
        
plt.figure(dpi=120)
plt.plot(angles/np.pi,distances)
plt.plot([0,0.4],[0,0],'--',color='k')
plt.plot([(angles[turn]+angles[turn-1])/(2*np.pi),(angles[turn]+angles[turn-1])/(2*np.pi)],[-120,350],'--',color='k')
plt.xlabel(r'Release angle $\theta_r$')
plt.ylabel('Distance')
plt.gca().xaxis.set_major_formatter(mticker.FormatStrFormatter('%.2f $\pi$'))
plt.grid()
plt.show()
```

We see that the trebuchet will fire forwards for any release angle $\theta_r < 0.17\pi$, while it will throw backwards for angle above that. This fits well with the plots earlier, where we could see that it changed somewhere between $0.15\pi$ and $0.2\pi$.

We are now going to plot the trajectories for different initial values $\theta_0$ and $\psi_0$.

```python
def init_diff(theta0,psi0):
    y = np.zeros(4)
    y[0] = theta0
    y[1] = theta0_der
    y[2] = psi0
    y[3] = psi0_der
    return y

arr2,t2 = RK4(0.001,5,F,init_diff(0.9*np.pi,psi0))

angs = [0,0.1,0.15,0.2]
lims = [0.001,0.01,0.01,0.001]
for i in range(len(angs)):
    sling,coords,start,end = trajectory(angs[i]*np.pi,lims[i],arr=arr2)
    plt.figure(dpi=120)
    plt.title(f'Release angle: {angs[i]}$\pi$ , '+r'$\theta_0=0.9\pi$')
    plt.plot(sling[0],sling[1])
    plt.plot(coords[:end,0],coords[:end,2])
    plt.plot(start[0],start[2],'o')
    plt.grid()
    plt.show()
```

With this new value of $\theta_0$ we can see that the trebuchet fires forward for all the plotted angles, and that the distance increases for increasing release angles.

```python
arr3,t3 = RK4(0.001,5,F,init_diff(theta0,0))

angs = [0,0.1,0.15,0.2]
lims = [0.001,0.01,0.01,0.01]
for i in range(len(angs)):
    sling,coords,start,end = trajectory(angs[i]*np.pi,lims[i],arr=arr3)
    plt.figure(dpi=120)
    plt.title(f'Release angle: {angs[i]}$\pi$ , $\psi_0=0$')
    plt.plot(sling[0],sling[1])
    plt.plot(coords[:end,0],coords[:end,2])
    plt.plot(start[0],start[2],'o')
    plt.grid()
    plt.show()
```

For $\psi_0=0$ the trebuchet mostly fires backwards, which comes from the fact that the angle $\psi$ will lag to much behind when $\theta=\theta_r$, which causes the projectile to be fired backwards.
